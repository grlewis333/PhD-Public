## <font color='gray'>Setup: </font> Generate data
- In general, data should consist of coordinates (x,y,z) each with a magnetisation vector (mx,my,mz)
- Code for interpolating finite element magnetisation data to finite domain representation is available separately
- Ultimately, this should be in the form X, Y, Z and MX,MY, MZ (where each is an (nxnxn) grid)

```python
def generate_tri_pris(n = 100, size_n = 1, mi = [1,0,0]):
    """ 
    Generate triangular prism data (with missing slice)
    
    Input:
    n = number of nodes in each dimension (nxnxn grid)
    size_n = length in nm of each node
    mi,ci = gradient/intercept of bounding line i for triangle
    mi = initial magnetisation vector 
    
    Output:
    X,Y,Z,MX,MY,MZ = Gridded coordinates, gridded magnetisation
    """
    
    # Define gradient/intercept of bounding lines
    m1, c1 = 5, 100
    m2, c2 = 0, -25
    m3, c3 = -0.6, 0
    
    # Generate x,y,z value
    xs = np.linspace(-n/2,n/2,int(n/size_n))
    ys = np.linspace(-n/2,n/2,int(n/size_n))
    zs = np.linspace(-n/2,n/2,int(n/size_n))
    
    X,Y,Z = np.meshgrid(xs,ys,zs,indexing='ij')

    # Assign magnetisation
    data = []
    for x in xs:
        for y in ys:
            for z in zs:
                if y < (m1*x+c1) and y > (m2*x + c2) and y < (m3*x + c3) and ((z >-20 and z<-10) or (z>0 and z<40)):
                    m = mi
                    data.append([x,y,z,m[0],m[1],m[2]])
                else:
                    m = [0,0,0]
                    data.append([x,y,z,m[0],m[1],m[2]])

    # Extract  magnetisations
    MX = np.take(data,3,axis=1)
    MY = np.take(data,4,axis=1)
    MZ = np.take(data,5,axis=1)

    MX = MX.reshape(len(xs),len(ys),len(zs))
    MY = MY.reshape(len(xs),len(ys),len(zs))
    MZ = MZ.reshape(len(xs),len(ys),len(zs))
    
    return X,Y,Z,MX,MY,MZ
```

## <font color='gray'>Setup: </font> Define 2D and 3D plotting functions

- 2D plot shows projection into the xy plane by averaging through z, and shows the Mx, My, and Mz components beneath
- 3D plot shows arrows matching the magnetisation, superimposed on a voxel plot that show where |M| is non-zero

```python
def plot_2d(X,Y,Z,MX,MY,MZ,s=5,size=0.1, width = 0.005, title=''):
    """
    Plot magnetisation data in 2D
    
    Input:
    x,y = 'Projected' 2D coordinates (nxn)
    u,v = 'Projected' 2D magnetisation (nxn)
    s = Quiver plot skip density
    size = Arrow length scaling
    width = Arrow thickness 
    
    Output:
    2D Plot of magnetisation:
    - Arrows show direction of M
    - Background color shows magnitude of M
    """
    # Project along z by averaging
    x_proj = np.mean(X,axis=2)
    y_proj = np.mean(Y,axis=2)
    z_proj = np.mean(Z,axis=2)
    u_proj = np.mean(MX,axis=2)
    v_proj = np.mean(MY,axis=2)
    w_proj = np.mean(MZ,axis=2)
    
    # Create figure
    fig = plt.figure(figsize=(6, 8))
    grid = plt.GridSpec(4, 3)
    ax1 = fig.add_subplot(grid[3, 0])
    ax2 = fig.add_subplot(grid[3, 1])
    ax3 = fig.add_subplot(grid[3, 2])
    ax4 = fig.add_subplot(grid[:3, :])
    
    # plot Mx
    pos =  ax1.imshow(np.flipud(u_proj.T), vmin=-1,vmax=1,cmap='RdBu'); 
    ax1.set_title('Mx',fontsize=13); ax1.set_xlabel('x',fontsize=14);ax1.set_ylabel('y',fontsize=14);

    # Plot My
    pos = ax2.imshow(np.flipud(v_proj.T), vmin=-1,vmax=1,cmap='RdBu');
    ax2.set_title('My',fontsize=13); ax2.set_xlabel('x',fontsize=14);ax2.set_ylabel('y',fontsize=14); 

    #Plot Mz
    pos = ax3.imshow(np.flipud(w_proj.T), vmin=-1,vmax=1,cmap='RdBu');
    ax3.set_title('Mz',fontsize=13); ax3.set_xlabel('x',fontsize=14);ax3.set_ylabel('y',fontsize=14); 
    fig.colorbar(pos,ax=ax3,fraction=0.046, pad=0.04)

    # Main arrow plot
    ax4.quiver(x_proj[::s,::s],y_proj[::s,::s],u_proj[::s,::s],v_proj[::s,::s],scale=1/size,pivot='mid',width=width)
    
    # Calculate vector magnitude
    magnitude = (u_proj**2 + v_proj**2)**0.5
    
    # Plot vector magnitude
    im1 = ax4.imshow(np.flipud(magnitude.T),vmin=0,vmax=1,cmap='Blues',
                     extent=(np.min(x_proj),np.max(x_proj),np.min(y_proj),np.max(y_proj)))
    
    # Add colorbar and labels
    clb = fig.colorbar(im1,ax=ax4,fraction=0.046, pad=0.04)
    ax4.set_xlabel('x / nm',fontsize=14)
    ax4.set_ylabel('y / nm',fontsize=14)
    ax4.set_title(title, fontsize= 16)
    for ax in [ax1,ax2,ax3]:
        ax.set_yticklabels([])
        ax.set_xticklabels([])
    
    plt.tight_layout()
```

```python
def plot_3d(X,Y,Z,MX,MY,MZ, s1=15, s2=3, title = '',el=70, az=-120,size_n=1):
    """ 
    Create 3D plot of magnetisation with quiver plot + volume render
    
    Inputs:
    X,Y,Z = Gridded 3D coordinates, each (nxnxn)
    u,v,w = Gridded 3D magnetisation data, each (nxnx)
    s1/s2 = Downsampling factor for quivers/voxels respectively
    title = Plot title
    el = elevation angle in z plane 
    az = azimuthal angle in xy plane
    """
    # Recreate the matplotlib fancy arrow patch
    # This is to plot arrows with filled triangles instead of line heads
    class Arrow3D(FancyArrowPatch):
        def __init__(self, xs, ys, zs, *args, **kwargs):
            FancyArrowPatch.__init__(self, (0, 0), (0, 0), *args, **kwargs)
            self._verts3d = xs, ys, zs

        def draw(self, renderer):
            xs3d, ys3d, zs3d = self._verts3d
            xs, ys, zs = proj3d.proj_transform(xs3d, ys3d, zs3d, renderer.M)
            self.set_positions((xs[0], ys[0]), (xs[1], ys[1]))
            FancyArrowPatch.draw(self, renderer)
    
    # Create 3D matplotlib plot
    fig = plt.figure(figsize=(6,6))
    ax = fig.gca(projection='3d')
    
    # Get downsampled array sizes
    r = int(np.shape(X)[0]/s1)
    n = int(np.shape(X)[0]/s2)
    
    # Plot each individual arrow by looping through x/y/z coordinates
    for ix in range(r):
        for iy in range(r):
            for iz in range(r):
                # Arrow goes from tail (X,Y,Z) to head (X+MX, Y+MY, Z+MZ)
                # Arrows are scaled down to match the size of the downsampled voxel array
                # and are then translated so that the centre of voxel/quiver arrays coincide
                # This makes it easier to plot the correct coordinates
                x,y,z = X[ix*s1,iy*s1,iz*s1],Y[ix*s1,iy*s1,iz*s1],Z[ix*s1,iy*s1,iz*s1]
                mx,my,mz = MX[ix*s1,iy*s1,iz*s1], MY[ix*s1,iy*s1,iz*s1],MZ[ix*s1,iy*s1,iz*s1]
                a = Arrow3D(np.array([x, x+mx*s1*size_n])/s2/size_n+n/2+s2/2/size_n,
                            np.array([y, y+my*s1*size_n])/s2/size_n+n/2+s2/2/size_n,
                            np.array([z, z+mz*s1*size_n])/s2/size_n+n/2+s2/2/size_n,
                            mutation_scale=20,lw=1, arrowstyle="-|>", color="k",alpha=1)
                ax.add_artist(a)

    
      # This code might be helpful if varying voxel colour is desired  
#     fc = np.ones((n,n,n,4))
#     for ix in range(n):
#         for iy in range(n):
#             for iz in range(n):
#                 mag = (MXr[ix*s2,iy*s2,iz*s2]**2+MYr[ix*s2,iy*s2,iz*s2]**2+MZr[ix*s2,iy*s2,iz*s2]**2)**.5
#                 if mag == 0:
#                     fc[ix,iy,iz] = [1,1,1,0]
#                 else:
#                     fc[ix,iy,iz] = [0,0,mag,alpha]
    
    # Plot voxel image based on magnitude of M
    # Note that the colors are given by [R, G, B, alpha]
    mag = (MX[::s2,::s2,::s2]**2+MY[::s2,::s2,::s2]**2+MZ[::s2,::s2,::s2]**2)**.5
    im = ax.voxels(mag, facecolors=[0,0,1,.1], edgecolor=[1,1,1,0.1])
    
    # Ticks are originally 0 to n (voxel matrix indices)
    # Here we rescale the tick labels to match the original coordinates
    ax.xaxis.set_ticks([0,n/2,n])
    ax.set_xticklabels([-int(np.shape(X)[0])*size_n/2,0,int(np.shape(X)[0])*size_n/2])
    ax.yaxis.set_ticks([0,n/2,n])
    ax.set_yticklabels([-int(np.shape(X)[0])*size_n/2,0,int(np.shape(X)[0])*size_n/2])
    ax.zaxis.set_ticks([0,n/2,n])
    ax.set_zticklabels([-int(np.shape(X)[0])*size_n/2,0,int(np.shape(X)[0])*size_n/2])
    
    # Set initial camera angle in degrees
    ax.view_init(elev=el, azim=az)
    
    # Add axis labels
    plt.xlabel('x / nm', fontsize=15)
    plt.ylabel('y / nm', fontsize=15)
    ax.set_zlabel('z / nm', fontsize=15)
    
    # Set background color to white
    ax.xaxis.pane.fill = False
    ax.yaxis.pane.fill = False
    ax.zaxis.pane.fill = False
    ax.xaxis.pane.set_edgecolor('w')
    ax.yaxis.pane.set_edgecolor('w')
    ax.zaxis.pane.set_edgecolor('w')
    plt.title(title, fontsize = 16)
    
    plt.tight_layout()
```

## <font color='gray'>Setup: </font> Define rotate functions
- This process is all encapsulated into the 'rotate_magnetisation' function
- The overall rotation is <i>intrinsic</i> (i.e. the rotation axes are 'attached' to the data, not extrinsic where the rotation axes are fixed in place (and correspond to the data axes), see <a href='https://en.wikipedia.org/wiki/Davenport_chained_rotations#Relationship_with_physical_motions'>here</a> for details)
- The algorithm first rotates the coordinates of the magnetisation data (using scipy's <a href='https://docs.scipy.org/doc/scipy/reference/generated/scipy.ndimage.rotate.html'>ndimage.rotate</a> function)
    - ndimage.rotate can perform <i>extrinsic</i> rotations of n-dimensional images about their axes - just specify the angle in degrees and the two axes defining the plane of rotation.
    - This function basically iterates a 2D rotation defined by
        \begin{bmatrix}
           \mathrm{cos}(\theta) & \mathrm{sin}(\theta) \\
       -\mathrm{sin}(\theta) & \mathrm{cos}(\theta) \\
      \end{bmatrix}
      over all planes parallel to the rotation plane. The transformation is applied to the coordinates of the image and then this result is interpolated back onto the original coordinate system for the end result.
    - By default, this uses 3rd order spline interpolation in which a cubic polynomial is fitted piecewise at each pixel to a subset of neighbouring pixels, in order to determine the value of the interpolated pixel. This process is lossy for all except 90$^{\circ}$ rotations.
    - The default also will fill any region outside the input data with zeros, though this can be changed to reflect/nearest/mirror/wrap behaviour.
    - Since this rotation is extrinsic, we need to perform the rotations in the opposite order to get intrinsic behaviour (see <a href='https://math.stackexchange.com/questions/1137745/proof-of-the-extrinsic-to-intrinsic-rotation-transform'>proof</a>.
- The magnetisation vector at each point is then rotated separately by applying a rotation matrix to each M vector. 
    - The rotation matrix is constructed by:
        \begin{gather}
        M_{x} = \begin{bmatrix}
           1 & 0 & 0 \\
           0 & \mathrm{cos}(\theta_x) & \mathrm{sin}(\theta_x) \\
           0 & -\mathrm{sin}(\theta_x) & \mathrm{cos}(\theta_x) \\
      \end{bmatrix}
      \end{gather}
      
      \begin{gather}
        M_{y} = \begin{bmatrix}
           \mathrm{cos}(\theta_y) & 0 & \mathrm{sin}(\theta_y) \\
           0 & 1 & 0 \\
           -\mathrm{sin}(\theta_y) &0& \mathrm{cos}(\theta_y)\\
      \end{bmatrix}
      \end{gather}
      
        \begin{gather}
        M_{z} = \begin{bmatrix}
           \mathrm{cos}(\theta_z) & -\mathrm{sin}(\theta_z) &0 \\
           \mathrm{sin}(\theta_z) & \mathrm{cos}(\theta_z) &0\\
                      0 & 0 & 1 \\
      \end{bmatrix}
      \end{gather}
      <center>$M_{rot} = M_{z} \cdot M_{y} \cdot M_{x} $</center>
    - In order to apply the rotation matrix, the magnetisation vectors must be extracted from their grids and formulated as coordinates. The matrix is then simply applied as $v_{rot} = M_{rot} \cdot v$, and the magnetisation components are then redistributed into their grids.

```python
def rotate_bulk(U,V,W,ax,ay,az):
    """ 
    Rotate magnetisation locations from rotation angles ax,ay,az 
    about the x,y,z axes (given in degrees) 
    
    NOTE: This implementation of scipy rotations is EXTRINSIC
    Therefore, to make it compatible with our intrinsic vector
    rotation, we swap the order of rotations (i.e. x then y then z)
    """
    # Due to indexing, ay needs reversing for desired behaviour
    ay = -ay
    
    U = ndimage.rotate(U,ax,reshape=False,axes=(1,2))
    U = ndimage.rotate(U,ay,reshape=False,axes=(2,0))
    U = ndimage.rotate(U,az,reshape=False,axes=(0,1))
    
    V = ndimage.rotate(V,ax,reshape=False,axes=(1,2))
    V = ndimage.rotate(V,ay,reshape=False,axes=(2,0))
    V = ndimage.rotate(V,az,reshape=False,axes=(0,1))
    
    W = ndimage.rotate(W,ax,reshape=False,axes=(1,2))
    W = ndimage.rotate(W,ay,reshape=False,axes=(2,0))
    W = ndimage.rotate(W,az,reshape=False,axes=(0,1))

    return U,V,W

def grid_to_coor(U,V,W):
    """ Convert gridded 3D data (3,n,n,n) into coordinates (n^3, 3) """
    coor_flat = []
    n = np.shape(U)[0]
    for ix in range(n):
        for iy in range(n):
            for iz in range(n):
                x = U[ix,iy,iz]
                y = V[ix,iy,iz]
                z = W[ix,iy,iz]
                coor_flat.append([x,y,z])
                
    return coor_flat

def coor_to_grid(coor_flat):
    """ Convert coordinates (n^3, 3) into gridded 3D data (3,n,n,n) """
    n = int(np.round(np.shape(coor_flat)[0]**(1/3)))
    
    x = np.take(coor_flat,0,axis=1)
    y = np.take(coor_flat,1,axis=1)
    z = np.take(coor_flat,2,axis=1)
    U = x.reshape((n,n,n))
    V = y.reshape((n,n,n))
    W = z.reshape((n,n,n))

    return U, V, W

def rotation_matrix(ax,ay,az):
    """ 
    Generate 3D rotation matrix from rotation angles ax,ay,az 
    about the x,y,z axes (given in degrees) 
    (Uses convention of rotating about z, then y, then x)
    """

    ax = ax * np.pi/180
    Cx = np.cos(ax)
    Sx = np.sin(ax)
    mrotx = np.array([[1,0,0],[0,Cx,-Sx],[0,Sx,Cx]])
    
    ay = ay * np.pi/180
    Cy = np.cos(ay)
    Sy = np.sin(ay)
    mroty = np.array([[Cy,0,Sy],[0,1,0],[-Sy,0,Cy]])
    
    az = az * np.pi/180
    Cz = np.cos(az)
    Sz = np.sin(az)
    mrotz = np.array([[Cz,-Sz,0],[Sz,Cz,0],[0,0,1]])
    
    mrot = mrotz.dot(mroty).dot(mrotx)
    
    return mrot

def rotate_vector(coor_flat,ax,ay,az):
    """ Rotates vectors by specified angles ax,ay,az 
    about the x,y,z axes (given in degrees) """
    
    # Get rotation matrix
    mrot = rotation_matrix(ax,ay,az)    

    coor_flat_r = np.zeros_like(coor_flat)
    
    # Apply rotation matrix to each M vector
    for i,M in enumerate(coor_flat):
        coor_flat_r[i] = mrot.dot(M)
    
    return coor_flat_r

def rotate_magnetisation(U,V,W,ax=0,ay=0,az=0):
    """ 
    Takes 3D gridded magnetisation values as input
    and returns them after an intrinsic rotation ax,ay,az 
    about the x,y,z axes (given in degrees) 
    (Uses convention of rotating about z, then y, then x)
    """
    # Rotate the gridded locations of M values
    Ub, Vb, Wb = rotate_bulk(U,V,W,ax,ay,az)
    
    # Convert gridded values to vectors
    coor_flat = grid_to_coor(Ub,Vb,Wb)
    
    # Rotate vectors
    coor_flat_r = rotate_vector(coor_flat,ax,ay,az)
    
    # Convert vectors back to gridded values
    Ur,Vr,Wr = coor_to_grid(coor_flat_r)
    
    # Set small values to 0
    # (In theory the magnitude of M in each cell should be Ms,
    #  so we can set magnitude lower than this to zero -
    #  typically python rounding errors lead to very small values,
    #  which it is useful to exclude here)
    mag_max = (np.max(U)**2+np.max(V)**2+np.max(W)**2)**0.5
    mag = (Ur**2+Vr**2+Wr**2)**.5
    for M in [Ur,Vr,Wr]:
        M[abs(M)<1e-5*mag_max] = 0
        M[mag<.6*mag_max] = 0
    
    return Ur,Vr,Wr
```

### Load data

```python
X,Y,Z,MX,MY,MZ = generate_tri_pris()
```

### Visualise input

```python
#2D
plot_2d(X,Y,Z,MX,MY,MZ, title = 'Input Magnetisation, projected along z')
```

```python
plot_3d(X,Y,Z,MX,MY,MZ)
```

### Rotate

```python
# To simulate electron microscope rotation:
# Simply set az = alpha, ay = beta and ax = 0
MXr,MYr,MZr = rotate_magnetisation(MX,MY,MZ,ax=0,ay=0,az=45)
```

### Visualise rotation

```python
# 2D
plot_2d(X,Y,Z,MXr,MYr,MZr, title = 'Rotated Magnetisation, projected along z')
```

```python
# 3D
plot_3d(X,Y,Z,MXr,MYr,MZr,title='Rotated magnetisation')
```

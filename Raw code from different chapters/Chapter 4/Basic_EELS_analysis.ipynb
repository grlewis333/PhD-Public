# Denoise

```python
# denoise the data
im.change_dtype('float')
im.decomposition(normalize_poissonian_noise = True, algorithm = 'svd', output_dimension = 30, tol = 1e-10)
im_denoised = im.get_decomposition_model()
```

```python
# remove any signal spikes, threshold ~100 works
im_denoised.spikes_removal_tool()
```

```python
# check scree plot
im_denoised.plot_explained_variance_ratio()
```

```python
# check prepared data
im_denoised.plot()
```

## Separate components by NNMF

```python
# Ensure that the nnmf algorithm is not fed negative data values
if im_denoised.data.min() < 0:
    im_denoised -= im_denoised.data.min()

# Choose number of dimensions
dim = 4

# Perform NMF decomposition
im_decomposed =  im_denoised.decomposition(normalize_poissonian_noise = True, algorithm = 'nmf', output_dimension = dim, tol = 5e-10,
                                           )
pic = im_denoised.plot_decomposition_loadings(per_row=1)
spec = im_denoised.plot_decomposition_factors()
```

## Background fit a particular component

```python
component1 = im_denoised.get_decomposition_factors().inav[0]
```

```python
# add elements and set microscope parameters
component1.set_microscope_parameters(beam_energy=60, convergence_angle=0, collection_angle=100)
component1.add_elements(elements=('O', 'C', 'Fe', 'Cu'))

# Initialise a model that will be fit to the specified edge
fm = component1.create_model()
fm.enable_fine_structure(edges_list=['O_K'])
fm.enable_fine_structure()
fm.smart_fit()
fm.fit_background()
fm.disable_background()
fm.resolve_fine_structure()

# view result
fm.plot()
```

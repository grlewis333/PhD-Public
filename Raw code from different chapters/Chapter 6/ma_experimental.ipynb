```python
def ft_thresh(im):
    # Invert
    im = 1/im
    
    # Binary threshold real space
    imax = np.max(im)
    imin = np.min(im)
    t = imax-.1*(imax-imin)
    im[im<t] = 0
    im[im>t] = 1
    
    # Erode 
    im = cv2.erode(im,cv2.getStructuringElement(cv2.MORPH_RECT, (8, 8)),iterations=1)
    
    # Hanning
    h0= np.hanning(np.size(ref,0))
    h1= np.hanning(np.size(ref,1))
    hann2d = np.sqrt(np.outer(h0,h1))
    im = im*hann2d

    # Fourier transform
    fft=np.log10(abs(np.fft.fftshift(np.fft.fft2(im))))**4
    fft = fft/np.max(fft)*255

    # threshold fourier space
    thresh = cv2.adaptiveThreshold(np.uint8(255-fft), 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 13, 3)
    thresh = 255 - thresh
    
    return thresh

def polar_projection(im):
    from skimage.transform import warp_polar
    from skimage.util import img_as_float
    
    # Transform into polar coordinate space
    image = img_as_float(im)
    image_polar = warp_polar(image)
    
    # Average out all radii
    polar_mean = np.mean(image_polar,axis=1)
    
    return polar_mean

def convolve_90deg(sig):
    # Create signal with deltas at 0 and 90 degrees
    sig_90 = np.zeros(90)
    sig_90[0]=1
    sig_90[89]=1
    
    # Convolve with polar signal to ensure peaks are 90 degree separated
    c = np.convolve(sig,sig_90)
    
    c=c[:180]
    
    return c

def draw_line(amax,cx,cy):
    sl = np.tan(np.radians(amax))
    x = np.arange(-10,10)
    y = sl*(x-cx) + cy
    #plt.plot(x,y, 'o')
    plt.axline((cx,cy), slope=sl, color='r', label='axline',alpha=.3,lw=10)
    
def unwrap(angs):
    angs = np.array(angs)
    r = angs%90
    for rep in range(3):
        for i in range(1,len(r)):
            if abs(r[i] - r[i-1]) > 45:
                r[i] = r[i]-90
                
    r = abs(r-r[0])
    return r

def find_nearest(a, a0):
    "Element in nd array `a` closest to the scalar value `a0`"
    idx = np.abs(a - a0).argmin()
    return a.flat[idx]
```

```python
def rotalign_search(angles,images,j=1,cw=True):
    """ 
    Returns the similary scores (COD) between each image in the stack
    with the jth previous image as a function of the rotation angle between them.
    """
    all_cods = []
    for i in range(j,len(images)):
        cods = []
        for a in angles:
            if cw == False:
                a=-a
            imrot=ndimage.rotate(images[i],-a,reshape=False)
            cod = ma.COD(ims[i-j],imrot)
            cods.append(cod)
        all_cods.append(cods)
        
    return all_cods

def find_best_alignment(all_cods):
    """ Returns argument of maximum similarity score """
    maxs = []
    for cod in all_cods:
        maxcod = np.argmax(cod)
        maxs.append(maxcod)
        
    return np.array(maxs)
```

```python
import matplotlib.lines as mlines
def line_detection_non_vectorized(image, edge_image, num_rhos=180, num_thetas=180, t_count=220):
    edge_height, edge_width = edge_image.shape[:2]
    edge_height_half, edge_width_half = edge_height / 2, edge_width / 2
    #
    d = np.sqrt(np.square(edge_height) + np.square(edge_width))
    dtheta = 180 / num_thetas
    drho = (2 * d) / num_rhos
    #
    thetas = np.arange(0, 180, step=dtheta)
    rhos = np.arange(-d, d, step=drho)
    #
    cos_thetas = np.cos(np.deg2rad(thetas))
    sin_thetas = np.sin(np.deg2rad(thetas))
    #
    accumulator = np.zeros((len(rhos), len(rhos)))
    #
    figure = plt.figure(figsize=(12, 12))
    subplot1 = figure.add_subplot(1, 4, 1)
    subplot1.imshow(image)
    subplot2 = figure.add_subplot(1, 4, 2)
    subplot2.imshow(edge_image, cmap="gray")
    subplot3 = figure.add_subplot(1, 4, 3)
    subplot3.set_facecolor((0, 0, 0))
    subplot4 = figure.add_subplot(1, 4, 4)
    subplot4.imshow(image)
    #
    for y in range(edge_height):
        for x in range(edge_width):
            if edge_image[y][x] != 0:
                edge_point = [y - edge_height_half, x - edge_width_half]
                ys, xs = [], []
                for theta_idx in range(len(thetas)):
                    rho = (edge_point[1] * cos_thetas[theta_idx]) + (edge_point[0] * sin_thetas[theta_idx])
                    theta = thetas[theta_idx]
                    rho_idx = np.argmin(np.abs(rhos - rho))
                    accumulator[rho_idx][theta_idx] += 1
                    ys.append(rho)
                    xs.append(theta)
                subplot3.plot(xs, ys, color="white", alpha=0.05)

    for y in range(accumulator.shape[0]):
        for x in range(accumulator.shape[1]):
            if accumulator[y][x] > t_count:
                rho = rhos[y]
                theta = thetas[x]
                a = np.cos(np.deg2rad(theta))
                b = np.sin(np.deg2rad(theta))
                x0 = (a * rho) + edge_width_half
                y0 = (b * rho) + edge_height_half
                x1 = int(x0 + 1000 * (-b))
                y1 = int(y0 + 1000 * (a))
                x2 = int(x0 - 1000 * (-b))
                y2 = int(y0 - 1000 * (a))
                subplot3.plot([theta], [rho], marker='o', color="yellow")
                subplot4.add_line(mlines.Line2D([x1, x2], [y1, y2]))

    subplot3.invert_yaxis()
    subplot3.invert_xaxis()

    subplot1.title.set_text("Original Image")
    subplot2.title.set_text("Edge Image")
    subplot3.title.set_text("Hough Space")
    subplot4.title.set_text("Detected Lines")
    plt.show()
    return accumulator, rhos, thetas
```

```python
def rotalign_search(rots,images,j=1,cw=True):
    """ 
    Returns the similary scores (COD) between each image in the stack
    with the jth previous image as a function of the rotation angle between them.
    """
    all_cods = []
    for i in range(j,len(images)):
        dif = rots[i]-rots[i-1]
        angles=np.linspace(dif-3,dif+3,11)
        print(angles)
        cods = []
        for a in angles:
            if cw == False:
                a=-a
            imrot=ndimage.rotate(images[i],-a,reshape=False)
            cod = ma.COD(images[i-1],imrot)
            cods.append(cod)
        all_cods.append(cods)
        
    return all_cods

def find_best_alignment(all_cods):
    """ Returns argument of maximum similarity score """
    maxs = []
    for cod in all_cods:
        maxcod = np.argmax(cod)
        maxs.append(maxcod)
        
    return np.array(maxs)

ac = rotalign_search(rots,ims,j=1,cw=True)
ms = find_best_alignment(ac)
```

```python
def draw_line(amax,cx,cy):
    sl = np.tan(np.radians(amax))
    x = np.arange(-10,10)
    y = sl*(x-cx) + cy
    #plt.plot(x,y, 'o')
    plt.axline((cx,cy), slope=sl, color='red', label='axline')
```

```python
def draw_line(angle,cx,cy):
    import math
    angle = angle%180
    angle = angle/180*np.pi
    length=np.min([cx,cy])
    endy = cy + length * np.sin(angle)
    endx = length * np.cos(angle)
    plt.plot([cx, endx], [cy, endy])
```

```python
def plot_point(point, angle, length):
    '''
    point - Tuple (x, y)
    angle - Angle you want your end point at in degrees.
    length - Length of the line you want to plot.

    Will plot the line on a 10 x 10 plot.
    '''
    import math
    # unpack the first point
    x, y = point

    # find the end point
    endy = y + length * math.sin(math.radians(angle))
    endx = length * math.cos(math.radians(angle))

    # plot the points
    fig = plt.figure()
    ax = plt.subplot(111)
    ax.set_ylim([0, 10])   # set the bounds to be 10, 10
    ax.set_xlim([0, 10])
    ax.plot([x, endx], [y, endy])

    fig.show()
        
plot_point((5,5),144,50)
```
